```python
import pymysql
import pandas as pd
```

```python
dbhost = 'localhost'
dbport = 3306
dbuser = 'root'
dbpasswd = ''
dbname = 'test'
```

```python
df = pd.read_csv('diabetes.csv', dtype=str)
df.head()
```

```python
df["chol"].fillna(0, inplace=True)
df["stab.glu"].fillna(0, inplace=True)
df["hdl"].fillna(0, inplace=True)
df["age"].fillna(0, inplace=True)
df["gender"].fillna('', inplace=True)
df["height"].fillna(0, inplace=True)
df["weight"].fillna(0, inplace=True)
df["frame"].fillna('', inplace=True)
df["bp.1s"].fillna(0, inplace=True)
df["bp.1d"].fillna(0, inplace=True)
```

```python
conn = pymysql.connect(host=dbhost, port=dbport, user=dbuser, passwd=dbpasswd, db=dbname)
conn.autocommit(1)
cursor = conn.cursor()

for index, row in df.iterrows():

    sql = "INSERT INTO diabetes (auto_id,id,chol,glu,hdl,age,gender,height,weight,frame,bps,bpd) "
    sql = sql + "VALUES (UUID(),'" + str(row["id"]) + "', '"
    sql = sql + str(row["chol"]) + "', '"
    sql = sql + str(row["stab.glu"]) + "', '"
    sql = sql + str(row["hdl"]) + "', '"
    sql = sql + str(row["age"]) + "', '"
    sql = sql + str(row["gender"]) + "', '"
    sql = sql + str(row["height"]) + "', '"
    sql = sql + str(row["weight"]) + "', '"
    sql = sql + str(row["frame"]) + "', '"
    sql = sql + str(row["bp.1s"]) + "', '"
    sql = sql + str(row["bp.1d"]) + "');"
    
    print(sql)
    cursor.execute(sql)
```
